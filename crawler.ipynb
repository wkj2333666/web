```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
header = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:140.0) Gecko/20100101 Firefox/140.0'}
netease_url = 'https://music.163.com'
response = requests.get(netease_url, params={}, headers=header)
response.encoding = 'utf-8'

netease_html = response.text
```

```python
netease_html
```

```python
netease_soup = BeautifulSoup(netease_html, 'html.parser')
```

```python
print(netease_soup.prettify())
```

```python
netease_soup.find_all('a')
```

```python
links = [link.get('href') for link in netease_soup.find_all('a')]
```

```python
toplist_url = r'https://music.163.com/discover/toplist?id=19723756'
toplist_html = requests.get(url=toplist_url, params={}, headers=header).text
toplist_soup = BeautifulSoup(toplist_html, 'html.parser')
file = toplist_soup.find_all('a', href=re.compile(r'/song\?id=\d'))
# file = toplist_soup.find_all('ul', class_='f-hide')
# file
with open('./toplist', 'w', encoding='utf-8') as toplist_file:
    for line in file:
        toplist_file.write(line.text)
        toplist_file.write('\n')
```
